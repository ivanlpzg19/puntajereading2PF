```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_students=pd.read_csv("clean_students_complete.csv")
data_students.head()
```

```python
#SE OBTIENEN EL PUNTAJE MÁS ALTO
pmax=data_students["reading_score"].max()
print("El mejor puntaje de reading_scores es de",pmax)
```

```python
prom=data_students[["gender","grade","school_name","reading_score"]] #se filtra información solo de reading_score
prom
          
```

```python
#SE FILTRAN SOLO A LOS PUNTAJES DE READING MÁS ALTOS
prommax=data_students[data_students["reading_score"]==data_students["reading_score"].max()]
prommax
```

```python
#SE REALIZA EN CONTEO DE LOS ALUMNOS POR ESCUELA CON EL MEJOR PUNTAJE
countmax=prommax.groupby(["grade","school_name"])["reading_score"].value_counts()
countmax
```

```python
#Se crea el gráfico que muestre el mejor puntaje por escuela y por grado

countmax.plot(kind="bar",color='#5F9EA0',figsize=(20,8))
plt.title("Alumnos con el mejor puntaje por escuela")
plt.ylabel("Alumnos")
```

```python
#Se crea el gráfico que muestre el mejor puntaje por escuela y por grado y por genero

countmax=prommax.groupby(["gender","grade","school_name"])["reading_score"].value_counts()
print("De los alumnos con mejor puntaje en cada escuela, por grado, la distribución se encuentra de la siguiente forma", countmax)

countmax.plot(kind="bar",color='#5F9EA0',figsize=(30,8))
plt.title("Generos por grado con mejores puntajes")
plt.ylabel("Alumnos")
```
